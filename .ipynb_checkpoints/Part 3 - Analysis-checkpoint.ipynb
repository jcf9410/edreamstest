```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import datetime
import calendar
```

```python
tweets_df = pd.read_csv('Tweets.csv')
delays_df = pd.read_csv('Delay_Airlines.csv')
weekdays = calendar.day_name

tweets_df['timestamp'] = pd.to_datetime(tweets_df['tweet_created'],format= '%Y-%m-%d %H:%M:%S')
```

## Airline popularity

```python
airline_popularity = tweets_df.groupby(['airline', 'airline_sentiment']).agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()

_airline_tweets = tweets_df.groupby('airline').agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()

_airline_tweets.rename(columns={
    'tweet_id': 'numtweets_by_airline',
    'retweet_count': 'retweet_count_by_airline'
}, inplace=True)
airline_popularity.rename(columns={'tweet_id': 'numtweets'}, inplace=True)

airline_popularity = airline_popularity.merge(_airline_tweets, on='airline')
airline_popularity['numtweets_percentage'] = 100 * airline_popularity[
    'numtweets'] / airline_popularity['numtweets_by_airline']
airline_popularity['retweet_count_percentage'] = 100 * airline_popularity[
    'retweet_count'] / airline_popularity['retweet_count_by_airline']

airline_popularity.sort_values(['airline', 'numtweets', 'retweet_count'],
                               ascending=[True, False, False],
                               inplace=True)

plotsize = (10, 20)
fig, axs = plt.subplots(4, figsize=plotsize)

sns.barplot(x="airline",
            y="numtweets",
            hue="airline_sentiment",
            data=airline_popularity,
            ax=axs[0]).set_title('Tweets by airline')
sns.barplot(x="airline",
            y="retweet_count",
            hue="airline_sentiment",
            data=airline_popularity,
            ax=axs[1]).set_title('Retweets by airline')
sns.barplot(x="airline",
            y="numtweets_percentage",
            hue="airline_sentiment",
            data=airline_popularity,
            ax=axs[2]).set_title('Tweets by airline %')
sns.barplot(x="airline",
            y="retweet_count_percentage",
            hue="airline_sentiment",
            data=airline_popularity,
            ax=axs[3]).set_title('Retweets by airline %')
plt.tight_layout()
```

## Tweets by sentiment and day

```python
tweets_df['weekday'] = tweets_df['timestamp'].apply(lambda x: x.weekday())

weekday_tweets = tweets_df.groupby(['weekday', 'airline_sentiment']).agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()
weekday_tweets.sort_values('weekday', inplace=True)
weekday_tweets['weekday'] = weekday_tweets['weekday'].apply(lambda x: weekdays[x])
weekday_tweets.rename(columns={'tweet_id': 'numtweets'}, inplace=True)

sns.set_style("darkgrid")
plotsize = (15, 15)
fig, axs = plt.subplots(2, figsize=plotsize)

sns.lineplot(x="weekday",
            y="numtweets",
            hue="airline_sentiment",
            data=weekday_tweets,
            ax=axs[0]).set_title('Tweets by weekday')
sns.lineplot(x="weekday",
            y="retweet_count",
            hue="airline_sentiment",
            data=weekday_tweets,
            ax=axs[1]).set_title('Retweets by weekday')
plt.tight_layout()
```

## Tweets by sentiment and hour

```python
tweets_df['hour'] = tweets_df['timestamp'].apply(lambda x: x.hour)

hourly_tweets = tweets_df.groupby(['hour', 'airline_sentiment']).agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()
hourly_tweets.sort_values('hour', inplace=True)
hourly_tweets.rename(columns={'tweet_id': 'numtweets'}, inplace=True)


sns.set_style("darkgrid")
plotsize = (15, 15)
fig, axs = plt.subplots(2, figsize=plotsize)


sns.lineplot(x="hour",
            y="numtweets",
            hue="airline_sentiment",
            data=hourly_tweets,
            ax=axs[0]).set_title('Tweets by hour')
sns.lineplot(x="hour",
            y="retweet_count",
            hue="airline_sentiment",
            data=hourly_tweets,
            ax=axs[1]).set_title('Retweets by hour')
plt.tight_layout()
```

## Correlation between tweets and cancelled flights.

```python
tweets_df['date'] = pd.to_datetime(tweets_df['tweet_created'],format= '%Y-%m-%d %H:%M:%S').dt.date
delays_df['FL_DATE'] = pd.to_datetime(delays_df['FL_DATE'],format= '%Y-%m-%d').dt.date
delays_df = delays_df[['FL_DATE', 'CANCELLED']]

tweets_grouped = tweets_df.groupby(['date', 'airline_sentiment']).agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()

# 2015-02-16 has missing Tweets data for full day (there are only a few tweets comparing with other days)
tweets_grouped = tweets_grouped[tweets_grouped['date'] > datetime.date(2015, 2, 16)]

delays_grouped = delays_df.groupby(['FL_DATE']).agg({
    'CANCELLED': 'sum'
}).reset_index()

_tweets_totals = tweets_df.groupby('date').agg({
    'tweet_id': 'count',
    'retweet_count': 'sum'
}).reset_index()
_tweets_totals.rename(columns={'tweet_id': 'total_numtweets', 'retweet_count': 'total_retweet_count'}, inplace=True)

tweets_grouped = tweets_grouped.merge(_tweets_totals, on='date')
tweets_grouped['percentage'] = 100* tweets_grouped['tweet_id'] / tweets_grouped['total_numtweets']

df = tweets_grouped.merge(delays_grouped, left_on='date', right_on='FL_DATE')
df.rename(columns={'tweet_id': 'numtweets', 'CANCELLED': 'cancelled_flights'}, inplace=True)
df.sort_values(['cancelled_flights', 'numtweets'], inplace=True)

df_neg = df[df['airline_sentiment'] == 'negative']

sns.set_style("darkgrid")
plotsize = (15, 5)
fig, axs = plt.subplots(1, figsize=plotsize)

sns.lineplot(x="cancelled_flights",
                y="percentage",
                data=df_neg,
                ax=axs).set_title(f'Tweets vs cancelled flights')    

plt.tight_layout()

# contar tweets por dia y sentimiento
# contar vuelos cancelados por dia
# ver relacion numtweets/vuelos cancelados por dia y sentimiento
```

```python
sentiments = ['negative', 'neutral', 'positive']

sns.set_style("darkgrid")
plotsize = (15, 10)
fig, axs = plt.subplots(2, figsize=plotsize)

df_log = df.copy()
df_log['numtweets'] = np.log(df_log['numtweets'])
df_log['cancelled_flights'] = np.log(df_log['cancelled_flights'])

sns.lineplot(x="cancelled_flights",
                y="numtweets",
                hue="airline_sentiment",
                data=df, #[df['airline_sentiment'] == sentiments[i]],
                ax=axs[0]).set_title(f' Tweets vs cancelled flights')   
sns.lineplot(x="cancelled_flights",
                y="numtweets",
                hue="airline_sentiment",
                data=df_log, #[df['airline_sentiment'] == sentiments[i]],
                ax=axs[1]).set_title(f' log Tweets vs log cancelled flights')   
plt.tight_layout()
```

```python
_df = df.copy()
_df['date'] = _df['date'] - datetime.timedelta(days=1)

sns.set_style("darkgrid")
plotsize = (15, 10)
fig, axs = plt.subplots(2, figsize=plotsize)

ax2 = axs[0].twinx() # applies twinx to ax2, which is the second y axis. 

sns.lineplot(x="date",
                y="cancelled_flights",
                data=df,
                ax=axs[0],
                color='r',
                legend='brief').set_title(f'Tweets vs cancelled flights')   

sns.lineplot(x="date",
                y="numtweets",
                hue='airline_sentiment',
                data=df,
                ax=ax2)

ax3 = axs[1].twinx() # applies twinx to ax2, which is the second y axis. 

sns.lineplot(x="date",
                y="cancelled_flights",
                data=df,
                ax=axs[1],
                color='r',
                legend='brief').set_title(f'Tweets vs cancelled flights, with lag of 1 day')   

sns.lineplot(x="date",
                y="numtweets",
                hue='airline_sentiment',
                data=_df,
                ax=ax3)


axs[0].set_ylabel('numtweets')
ax2.set_ylabel('cancelled_flights')
axs[1].set_ylabel('numtweets')
ax3.set_ylabel('cancelled_flights')
plt.tight_layout()
```

```python
print(f"correlation for all data: {df['cancelled_flights'].corr(df['numtweets'])}")

for s in ['negative', 'neutral', 'positive']:
    _df = df[df['airline_sentiment'] == s]
    print(f"correlation for {s} tweets: {_df['cancelled_flights'].corr(_df['numtweets'])}")

delays_grouped = delays_df.groupby(['FL_DATE']).agg({
    'CANCELLED': 'sum'
}).reset_index()
          
delays_grouped['FL_DATE'] = delays_grouped['FL_DATE'] - datetime.timedelta(days=1)

_df = tweets_grouped.merge(delays_grouped, left_on='date', right_on='FL_DATE')
_df.rename(columns={'tweet_id': 'numtweets', 'CANCELLED': 'cancelled_flights'}, inplace=True)
_df.sort_values(['cancelled_flights', 'numtweets'], inplace=True)
          
print(f"correlation for all data, with lag: {_df['cancelled_flights'].corr(_df['numtweets'])}")

for s in ['negative', 'neutral', 'positive']:
    __df = _df[_df['airline_sentiment'] == s]
    print(f"correlation for {s} tweets, with lag: {__df['cancelled_flights'].corr(__df['numtweets'])}")
          
##################################

_df_log = df_log.copy()
print(f"log correlation for all data: {df_log['cancelled_flights'].corr(df_log['numtweets'])}")

for s in ['negative', 'neutral', 'positive']:
    _df = df_log[df_log['airline_sentiment'] == s]
    print(f"log correlation for {s} tweets: {_df['cancelled_flights'].corr(_df['numtweets'])}")

delays_grouped = delays_df.groupby(['FL_DATE']).agg({
    'CANCELLED': 'sum'
}).reset_index()
          
delays_grouped['FL_DATE'] = delays_grouped['FL_DATE'] - datetime.timedelta(days=1)

_df = tweets_grouped.merge(delays_grouped, left_on='date', right_on='FL_DATE')
_df.rename(columns={'tweet_id': 'numtweets', 'CANCELLED': 'cancelled_flights'}, inplace=True)
_df.sort_values(['cancelled_flights', 'numtweets'], inplace=True)
          
_df_log['numtweets'] = np.log(_df['numtweets'])
_df_log['cancelled_flights'] = np.log(_df['cancelled_flights'])
          
print(f"log correlation for all data, with lag: {_df_log['cancelled_flights'].corr(_df_log['numtweets'])}")

for s in ['negative', 'neutral', 'positive']:
    __df = _df_log[_df_log['airline_sentiment'] == s]
    print(f"log correlation for {s} tweets, with lag: {__df['cancelled_flights'].corr(__df['numtweets'])}")
          
          
```

```python
import statsmodels.api as sm

X = df[df['airline_sentiment'] == 'negative']['cancelled_flights']
y = df[df['airline_sentiment'] == 'negative']['numtweets']

model = sm.OLS(y, X)
results = model.fit()

print("Normal values")
print(f"p-value: {results.pvalues}")
print(results.summary())

fig, ax = plt.subplots(figsize=(15,5))

ax.plot(X, y, 'o', label="data")
ax.plot(X, results.fittedvalues, 'r--.', label="OLS")
ax.legend(loc='best');
```
